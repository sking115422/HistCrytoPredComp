```python
import pandas as pd
import os 
```

```python
if os.path.isdir('../raw_data') == False:
    os.mkdir('../raw_data')
```

```python
data = pd.read_csv ('train.csv')
```

```python
data['Time'] = pd.to_datetime(data['timestamp'], unit='s')
```

```python
asset_details = pd.read_csv('asset_details.csv')
```

```python
data.head(10)
```

```python
asset_details.head(15)
```

```python
asset_list = []

for each in asset_details['Asset_ID']:
    print(each)
    asset_list.append( pd.DataFrame(data.loc[data['Asset_ID'] == each ]) )
    
```

```python
count = 0

for each in asset_list:
    print (count, ")   ", asset_list[count].shape[0])
    count = count + 1
    
```

```python
BTC_data = pd.DataFrame(asset_list[1])
ETH_data = pd.DataFrame(asset_list[6])
```

```python
BTC_data.tail(20)
```

```python
start_date_train = '2018-01-01 00:00:00'
end_date_train = '2021-01-01 00:00:00'
```

```python
start_date_test = '2021-01-01 00:00:00'
end_date_test = '2021-09-01 00:00:00'
```

```python
### MIN TEST and TRAIN RAW DATA
```

```python
#BTC MIN Train

mask = (BTC_data['Time'] > start_date_train) & (BTC_data['Time'] <= end_date_train)

BTC_data_trim_train_min = pd.DataFrame(BTC_data.loc[mask])

BTC_data_trim_train_min.drop('Target', inplace=True, axis=1)

BTC_data_trim_train_min = BTC_data_trim_train_min.reset_index(drop=True)

BTC_data_trim_train_min.to_csv('../raw_data/BTC_min_train.csv')
```

```python
#BTC MIN Test

mask = (BTC_data['Time'] > start_date_test) & (BTC_data['Time'] <= end_date_test)

BTC_data_trim_test_min = pd.DataFrame(BTC_data.loc[mask])

BTC_data_trim_test_min.drop('Target', inplace=True, axis=1)

BTC_data_trim_test_min = BTC_data_trim_test_min.reset_index(drop=True)

BTC_data_trim_test_min.to_csv('../raw_data/BTC_min_test.csv')
```

```python
### HOUR TEST and TRAIN RAW DATA
```

```python
#BTC HOUR Train

#BTC DAY Train

mask = (BTC_data['Time'] > start_date_train) & (BTC_data['Time'] <= end_date_train)

BTC_data_trim_train_hour = pd.DataFrame(BTC_data.loc[mask])

BTC_data_trim_train_hour.drop('Target', inplace=True, axis=1)

BTC_data_trim_train_hour = BTC_data_trim_train_hour.reset_index(drop=True)


vol_list = []
count_list = []
sum_vol = 0
sum_count = 0
for each in range(0, BTC_data_trim_train_hour.shape[0]):
    if BTC_data_trim_train_hour['timestamp'].values[each] % 3600 != 0:
        sum_vol = sum_vol + BTC_data_trim_train_hour['Volume'].values[each]
        sum_count = sum_count + BTC_data_trim_train_hour['Count'].values[each]
    else:
        vol_list.append(sum_vol)
        count_list.append(sum_count)
        sum_vol = 0
        sum_count = 0
    


temp = BTC_data_trim_train_hour.loc[BTC_data_trim_train_hour['timestamp'] % 3600 == 0]

BTC_data_trim_train_hour = temp

BTC_data_trim_train_hour = BTC_data_trim_train_hour.reset_index(drop=True)

BTC_data_trim_train_hour.drop(columns=['Volume', 'Count'])

BTC_data_trim_train_hour['Volume'] = vol_list
BTC_data_trim_train_hour['Count'] = count_list

BTC_data_trim_train_hour

BTC_data_trim_train_hour.to_csv('../raw_data/BTC_hour_train.csv')
```
